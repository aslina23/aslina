```python
df = pd.read_csv("data.csv")
df
```

```python
cars_grouped_make = df.groupby('Make')

for make, group in cars_grouped_make:
    print(make)
    print(group [['Make','Model','Year','MSRP']])
```

```python
# find the most expensive and cheapest, mean price for each make

cars_grouped_make = df[['Vehicle Style', 'Market Category','MSRP']
                      ].groupby([df['Make'],df['Model'],df['Year']])

for make,group in cars_grouped_make:
    print(make)
    print(f"most expensive : ${group['MSRP'].max()}")
    print(f"cheapest       : ${group['MSRP'].min()}")
    print(f"mean           : ${group['MSRP'].mean()}")
    print(" ")
```

```python
#find the most expensive car

df.sort_values(by = 'MSRP', ascending = False).iloc[0] #iloc[0] return only the most expensive
```
